For Multiplication:
Each element in the resulting (n,m) matrix is obtained by taking the dot product of a row from the (n,k) matrix and a column from the (k,m) matrix.
So total number of multiplication is k times number of elements(n*m) that is equal to nmk .
For Addition,
After performing the k multiplications for each element in the resulting matrix, you need to sum these products.
For each element, you need to perform kâˆ’1 additions. Therefore total number of additions are k-1 times n*k that is equal to nm(k-1) .

```python
# Question 2

import time
import numpy as np

# Define your matrices here
A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
B = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]


# List of list method
start = time.time()
AxB = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

ar = len(A)
ac = len(A[0])
br = len(B)
bc = len(B[0])

if ac == br:
    for i in range(ar):
        for j in range(bc):
            for k in range(br):
                AxB[i][j] += A[i][k] * B[k][j]
    print("Result  using list of list method:")
    for row in AxB:
        print(row)
else:
    print("Matrix Multiplication is not possible")

end = time.time()
print("Time Taken by the list of list method =", end - start)

# Numpy method
start = time.time()
P = np.array(A)
Q = np.array(B)
PxQ = np.dot(P, Q)

print("\nResult  using numpy method:")
for row in PxQ:
    print(row)

end = time.time()
print("Time Taken by the numpy method =", end - start)

```

As seen above, it is clear that numpy method takes less time than list of lists method. This is because Numpy makes use of fragmentation i.e. it divided the task into fragments and then runs all the fragments simultaneously

Question 3:

To find median of an array usking iterative method, we need to make multiple passes and hence the time complexity by this method would be n square.
A better method would be Quickselect Algorithm as shown below

```python
import time
import numpy as np
import random

# Method 1: Sequential Passes
def find_median_sequential(arr):
    n = len(arr)
    for i in range((n // 2) + 1):
        max_val = max(arr)
        if i == (n // 2):
            return max_val
        arr.remove(max_val)

# Method 2: Quickselect
def quickselect(arr, k):
    if len(arr) == 1:
        return arr[0]

    pivot = random.choice(arr)
    lows = [el for el in arr if el < pivot]
    highs = [el for el in arr if el > pivot]
    pivots = [el for el in arr if el == pivot]

    if k < len(lows):
        return quickselect(lows, k)
    elif k < len(lows) + len(pivots):
        return pivots[0]
    else:
        return quickselect(highs, k - len(lows) - len(pivots))

def find_median_quickselect(arr):
    n = len(arr)
    return quickselect(arr, n // 2)


def test_methods(arr):

    arr_copy1 = arr.copy()
    arr_copy2 = arr.copy()

    start = time.time()
    median_seq = find_median_sequential(arr_copy1)
    end = time.time()
    print(f"Median by sequential method: {median_seq}, Time taken: {end - start:.6f} seconds")

    start = time.time()
    median_quick = find_median_quickselect(arr_copy2)
    end = time.time()
    print(f"Median by quickselect method: {median_quick}, Time taken: {end - start:.6f} seconds")

    start = time.time()
    median_np = np.median(arr)
    end = time.time()
    print(f"Median by numpy method: {median_np}, Time taken: {end - start:.6f} seconds")

# Example usage
arr = [random.randint(1, 10000) for _ in range(1000)]
test_methods(arr)
```

```python
# Question 6

import sympy as sp
x, y = sp.symbols('x y')
func = (x**2)*y + (y**3)*sp.sin(x)
gradx = sp.diff(func,x)
grady = sp.diff(func,y)

print("The gradient of the following function with respect to x is ",gradx)
print("The gradient of the following function with respect to y is ",grady)
```

Broadcasting is a feature in Numpy that allows the user to perform various operations between arrays of different shapes and size.

```python
#Example
import numpy as np

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

vector = np.array([10, 20, 30])

result = matrix + vector

print("Matrix:")
print(matrix)

print("\nVector:")
print(vector)

print("\nResult of adding matrix and vector (broadcasting):")
print(result)
```

```python
#Question 13
import numpy as np
def argmin(arr):
    if not isinstance(arr, np.ndarray):
        raise ValueError("Array is not a Numpy array")
    if arr.size == 0:
        raise ValueError(" Array is an empty array")

    index = 0
    value = arr[0]
    for i in range(0,arr.size):
        if arr[i]<value:
            value = arr[i]
            index = i
    return index
a = np.array([3,2,1,5,2,8,4,3,3])
ans = argmin(a)
print("Index of the minimum element using function defined by me = ", ans)
np_result = np.argmin(a)
print("Index of the minimum element using the numpy function = ", np_result)
```
